## Ejercicio 1: Creación de una Clase Básica
Crea una clase llamada `Libro` que tenga los siguientes atributos:
- título
- autor
- año_publicacion
- disponible (booleano)

Implementa los métodos:
- `prestar()`: cambia disponible a False
- `devolver()`: cambia disponible a True
- `info()`: muestra toda la información del libro

```python
class Libro:
    def __init__(self,titulo,autor,year_publicacion,disponible):
        self.titulo = titulo
        self.autor = autor
        self.year_publicacion = year_publicacion
        self.disponible = disponible 
        pass
    
    def prestar(self):
        self.disponible = False
    
    def devolver(self):
        self.disponible = True
        
    def info(self):
        print(self.titulo)
        print(self.autor)
        print(self.year_publicacion)
        print(f"Esta disponible: {self.disponible}")
        
    
libro1 = Libro("Harry Potter","JK Rowling","2004",True)
libro2 = Libro("Geronimo Stilton 2","Geronimo Stilton","2011",False)
libro2.devolver()
libro2.info()
print("\n")
libro1.prestar()
libro1.info()
        
```

## Ejercicio 2: Herencia
Crea una clase `Vehiculo` con atributos básicos como `marca`, `modelo` y `año`.
Luego crea dos clases que hereden de `Vehiculo`:
- `Coche` (con atributo adicional `num_puertas`)
- `Moto` (con atributo adicional `cilindrada`)

Cada clase debe tener un método `mostrar_info()` que muestre sus atributos.

```python
class Vehiculo:
    
    def __init__(self,marca,modelo,year):
        self.marca = marca
        self.modelo = modelo
        self.year = year
        pass

class Coche(Vehiculo):
    
    def __init__(self,marca,modelo,year,num_puertas):
        super().__init__(marca,modelo,year)
        self.num_puertas = num_puertas
        pass
        
    def mostrar_info(self):
        print(self.marca)
        print(self.modelo)
        print(self.year)
        print(f"Tiene {self.num_puertas} puertas")
        
    
class Moto(Vehiculo):
    
    def __init__(self, marca, modelo, year,cilindradas):
        super().__init__(marca, modelo, year)
        self.cilindradas = cilindradas
        pass
    
    def mostrar_info(self):
        print(self.marca)
        print(self.modelo)
        print(self.year)
        print(f"Tiene {self.cilindradas} cilindradas")
        

coche1 = Coche("BMW","IX3","2024","4")
moto1 = Moto("Aprilia","RS 660","2024","125")

coche1.mostrar_info()
print("\n")
moto1.mostrar_info()

    
```

## Ejercicio 3: Encapsulamiento
Crea una clase `CuentaBancaria` con:
- Atributos privados: `__saldo` y `__num_cuenta`
- Métodos:
  - `depositar(cantidad)`
  - `retirar(cantidad)`
  - `consultar_saldo()`

Asegúrate de que no se pueda retirar más dinero del disponible.

```python
class CuentaBancaria:
    
    def __init__(self, __saldo,__num_cuenta):
        self.__saldo = __saldo
        self.__num_cuenta = __num_cuenta
    
    def depositar(self,cantidad):
        self.__saldo += cantidad
     
    
    def retirar(self,cantidad):
        if(cantidad<=self.__saldo):
         self.__saldo -= cantidad
         print(f"Has retirado {cantidad}$")
        else:
            print("Saldo insuficiente")
        
    def consultar_saldo(self):
        print(f"{self.__saldo}$")
        

cuenta1 = CuentaBancaria(50,45678)
cuenta1.depositar(30)
cuenta1.consultar_saldo()
cuenta1.retirar(90)
```
